```python
df = pd.read_csv("SampleSuperstore.csv")
```

```python
print("ðŸ”¹ 5 Baris Pertama Dataset:")
display(df.head())

print("\nðŸ”¹ Informasi Dataset:")
print(df.info())

print("\nðŸ”¹ Statistik Ringkas:")
display(df.describe())
```

```python
print("\nðŸ”¹ Nilai Kosong per Kolom:")
print(df.isnull().sum())

print("\nðŸ”¹ Jumlah Data Duplikat:", df.duplicated().sum())
```

```python
df.drop_duplicates(inplace=True)
```

```python
print("\nðŸ”¹ Jumlah Data Setelah Pembersihan:", df.shape)
print("\nðŸ”¹ Kolom yang Tersedia:", df.columns.tolist())
```

```python
category_perf = df.groupby('Category')[['Sales','Profit']].sum().sort_values(by='Profit', ascending=False)
print("\nðŸ”¹ Penjualan & Profit per Category:")
display(category_perf)
```

```python
region_perf = df.groupby('Region')['Profit'].sum().sort_values(ascending=False)
print("\nðŸ”¹ Profit per Region:")
display(region_perf)
```

```python
subcat_perf = df.groupby('Sub-Category')[['Sales','Profit']].sum().sort_values(by='Profit', ascending=False)
print("\nðŸ”¹ Sub-Category dengan Profit Tertinggi:")
display(subcat_perf.head(10))
```

```python
plt.figure(figsize=(7,4))
sns.barplot(x='Category', y='Profit', data=df, palette='viridis')
plt.title('Profit per Category')
plt.show()

plt.figure(figsize=(7,4))
sns.barplot(x='Region', y='Sales', data=df, palette='plasma')
plt.title('Sales per Region')
plt.show()

plt.figure(figsize=(10,5))
sns.barplot(x='Sub-Category', y='Profit', data=df, palette='coolwarm', order=subcat_perf.index)
plt.xticks(rotation=75)
plt.title('Profit per Sub-Category')
plt.show()
```

```python
plt.figure(figsize=(6,4))
sns.scatterplot(x='Sales', y='Profit', data=df, color='blue')
plt.title('Korelasi antara Sales dan Profit')
plt.show()

corr = df[['Sales','Profit']].corr()
print("\nðŸ”¹ Korelasi antara Sales dan Profit:")
display(corr)
```

```python
loss_products = df[df['Profit'] < 0]
print("\nðŸ”¹ Contoh Produk yang Rugi:")
display(loss_products.head(5))
```
